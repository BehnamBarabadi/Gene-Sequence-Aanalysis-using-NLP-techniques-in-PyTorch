### Let's find the number of pfams in Responder and NonResponders

```python
len(df)
```

```python
from collections import Counter

def counters(df):
    
    responder_pfam_counts = Counter()
    non_responder_pfam_counts = Counter()
    total_pfams_counts =  Counter()
    cols = df.columns[2:]
    
    for index, row in df.iterrows():
        if row.Response == 'Responder':
            for col in cols:
                if df.loc[index, col] == 1:                            
                    responder_pfam_counts[col] += 1
                    total_pfams_counts[col] += 1
        
        else:
            for col in cols:
                if df.loc[index, col] == 1:                            
                    non_responder_pfam_counts[col] += 1
                    total_pfams_counts[col] += 1
                
    return responder_pfam_counts, non_responder_pfam_counts, total_pfams_counts
```

```python
pfam_counter = counters(df)
```

```python
# Most common pfams in NonResponders
pfams_in_responders = pfam_counter[0]
print("number of pfams in Responder: ",len(pfams_in_responders))
pfams_in_responders.most_common(50)
```

```python
# Most common pfams in NonResponders
pfams_in_non_responders = pfam_counter[1]
print("number of pfams in NonResponder: ",len(pfams_in_non_responders))
pfams_in_non_responders.most_common(50)
```

```python
# Most common pfams in NonResponders
total_pfams = pfam_counter[2]
total_pfams.most_common(50)
```

```python
pfams_in_responders.most_common(10)
```

### pos_neg_ratios

```python
pos_neg_ratios = Counter()
# Calculate the ratios of positive and negative uses of the most common words
# Consider words to be "common" if they've been used at least 100 times
for term,cnt in list(total_pfams.most_common()):
#     if(cnt > 1):
    pos_neg_ratio = pfams_in_responders[term] / float(pfams_in_non_responders[term]+1)
    pos_neg_ratios[term] = pos_neg_ratio
```

```python
pos_neg_ratios
```

```python
# pfams most frequently seen in a review with a "Responder" label
# pfams with highest ratio of Responder to NonResponder
pos_neg_ratios.most_common(10)
```

```python
# pfams most frequently seen in a review with a "NonResponder" label
pos_neg_ratios.most_common()[:-31:-1]
```

### neg_pos_ratios

```python
neg_pos_ratios = Counter()
# Calculate the ratios of positive and negative uses of the most common words
# Consider words to be "common" if they've been used at least 100 times
for term,cnt in list(total_pfams.most_common()):
#     if(cnt > 1):
    neg_pos_ratio = pfams_in_non_responders[term] / float(pfams_in_responders[term]+1)
    neg_pos_ratios[term] = neg_pos_ratio
```
